```python
import numpy as np

# Create synthetic data: 100 samples, 3 features each
np.random.seed(42)
X = np.random.randn(100, 3)
# True weights and bias we want our model to discover
true_w = np.array([2.5, -1.5, 0.5])
true_b = 3.0
y = np.dot(X, true_w) + true_b + np.random.randn(100) * 0.1
```

```python
w = np.zeros(3)
b = 0.0
```

```python
def predict(X, w, b):
    y_hat = np.dot(X, w) + b
    return y_hat
```

```python
def compute_cost(X, y, w, b):
    """Calculates the Mean Squared Error (MSE) cost function."""
    m = len(y)
    y_hat = predict(X, w, b)
    cost = (1 / (2 * m)) * np.sum((y_hat - y) ** 2)
    return cost
```

```python
def gradient_descent(X, y, w, b, alpha, num_iters):
    """Optimizes weights and bias using batch gradient descent."""
    m = len(y)
    cost_history = []

    for i in range(num_iters):
        y_hat = predict(X, w, b)

        # Compute partial derivatives simultaneously
        dj_dw = np.dot(X.T, (y_hat - y)) / m
        dj_db = np.sum(y_hat - y) / m

        # Update parameters via gradient steps
        w = w - (alpha * dj_dw)
        b = b - (alpha * dj_db)

        # Track cost convergence
        if i % 100 == 0:
            cost = compute_cost(X, y, w, b)
            cost_history.append(cost)
            print(f"Iteration {i:04d} | Cost: {cost:.6f}")

    return w, b, cost_history
```

```python
# Hyperparameters
ALPHA = 0.1
NUM_ITERS = 1000

# Execute Optimization Loop
final_w, final_b, cost_history = gradient_descent(X, y, w, b, ALPHA, NUM_ITERS)

print("\n=================== MODEL EVALUATION ===================")
print(f"Target Weights: {true_w} -> Optimized Weights: {np.round(final_w, 4)}")
print(f"Target Bias:    {true_b}             -> Optimized Bias:    {final_b:.4f}")
print("========================================================")
```
