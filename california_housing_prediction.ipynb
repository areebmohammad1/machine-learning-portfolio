# California Housing Price Prediction Pipeline
This project implements an end-to-end regression workflow using Scikit-Learn to predict median house values in California.

### Core Architecture:
* **Data Source:** California Housing Dataset (fetched via `sklearn.datasets`)
* **Feature Processing:** Automated feature matrix ($X$) and target vector ($y$) isolation with an 80/20 train/test split.
* **Model Engine:** `RandomForestRegressor` optimized to evaluate continuous numerical trends.
* **Evaluation Metric:** Mean Absolute Error (MAE)

```python
# Cell 1: Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# Cell 2: Fetch and build the DataFrame
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df["target"] = housing["target"]
```

```python
# Cell 3: Split into X and y
X = housing_df.drop("target", axis=1)
y = housing_df["target"]
```

```python
# Cell 4: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Cell 5: Check shapes
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Import algorithm/estimator
from sklearn.linear_model import Ridge

#Setup random seed
np.random.seed(42)

#Instantiate and fit the model (on the training set)
model = Ridge()
model.fit(X_train, y_train)

#Check the score of the model (on the test set)
model.score(X_test, y_test)
```

What if 'Ridge' didn't work or the score didn't fit our needs?

Well, we could always try a different model...

How about we try an ensemble model (an esemble is a combination of smaller models to try and make better predictions than just a single model)

```python
housing_df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Setup random seed
np.random.seed(42)

# Create model instance
model = RandomForestRegressor()

# Fit the model to the data
model.fit(X_train, y_train)

# Check the baseline score (R^2)
print(f"Model Score (R^2): {model.score(X_test, y_test)}")

# Make predictions
y_preds = model.predict(X_test)
```
